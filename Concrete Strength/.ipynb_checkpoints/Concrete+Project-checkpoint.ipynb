# Dealing with Outliers of the dataset

```python
column_outliers = ['cement', 'ash', 'slag', 'water', 'superplastic', 'coarseagg', 'fineagg', 'age', 'strength']

for i in column_outliers:
    dp75,dp25 = np.percentile(concrete_df1[i], [75,25])
    IQR = dp75 - dp25
    Up_limit = dp75 + 1.5 * IQR
    Low_limit = dp25 - 1.5 * IQR
    median = concrete_df1[i].median()
    print('Count of outliers of', i)
    print('Number of outliers in the upper range is ', concrete_df1[concrete_df1[i] > Up_limit][i].shape[0])
    print('Number of outliers in the lower range is ', concrete_df1[concrete_df1[i] < Low_limit][i].shape[0])
  #  concrete_df1.loc[(concrete_df1[i] > Up_limit),i] = median
  #  concrete_df1.loc[(concrete_df1[i] < Low_limit),i] = median
    
    print('*' * 100)
```


- Outliers of  attributes are better to be replaced with nearest values like (mean + 2sigma) for upper range outliers and (mean - 2sigma) for lower range outliers

```python
# Replacing the outliers with suitable values for the independent variables only 

column_outliers = [ 'slag', 'water', 'superplastic', 'fineagg', 'age']

for i in column_outliers:
    dp75,dp25 = np.percentile(concrete_df1[i], [75,25])
    IQR = dp75 - dp25
    Up_limit = dp75 + 1.5 * IQR
    Low_limit = dp25 - 1.5 * IQR
    mean = concrete_df1[i].mean()
    std = concrete_df1[i].std()
    print('Number of outliers in the upper range is ', concrete_df1[concrete_df1[i] > Up_limit][i].shape[0])
    print('Number of outliers in the lower range is ', concrete_df1[concrete_df1[i] < Low_limit][i].shape[0])
    concrete_df1.loc[(concrete_df1[i] >Up_limit),i] = mean + 2 * std
    concrete_df1.loc[(concrete_df1[i] < Low_limit),i] = mean - 2 * std
    msg = " Upper outliers are replaced by %.2f and Lower outliers if any are replaced by %.2f" % ((mean + 2 *std), (mean - 2 * std))
    print(msg)
    print('All the outliers are replaced of ', i)
    print('*' * 100)
    
```

# Feature importance using the P value  

```python

# Thy hyposthesis is that the predicted value( cement strength) is not dependent on the independent variable
# If the P value is greater than the threshold fixed at 0.05, then the hypothesis is true otherwise rejected
```

```python
selected_columns = concrete_df2.columns
```

```python
selected_columns = selected_columns[0:10]
```

```python

import statsmodels.formula.api as sm
def BackwardElimination(x, Y, sl, columns):
    numVars = len(x[0])
    print(numVars)
    for i in range(0, numVars):
        regressor_OLS = sm.OLS(Y, x).fit()
        maxVar = max(regressor_OLS.pvalues).astype(float)
        print(maxVar)
        if maxVar > sl:
            for j in range(0, numVars - i):
                if (regressor_OLS.pvalues[j].astype(float) == maxVar):
                    x = np.delete(x, j, 1)
                    columns = np.delete(columns, j)
                    
    print(regressor_OLS.summary())
    return x, columns
SL = 0.05
data_modeled, selected_columns = BackwardElimination(concrete_df2.iloc[:,:10].values, concrete_df2.iloc[:,10].values, SL, selected_columns)
selected_columns
```

 
- P-value feature elimination process gives a new array of features without CARatio as attribute.

# CALCULATION OF ACCURACY CONFIDENT INTERVAL AT 95% CONFIDENCE

```python
# CONFIDENT INTERVAL OF ACCURACY OF DIFFERENT MODELS IS DONE BY BOOTSTRAPPING METHOD 
```

- CONFIDENCE INTERVAL OF ACCURACY FOR POLYNOMIAL MODEL

```python
# Initiating Models
clfLR = LinearRegression()
clfridge = grid_search_ridge_p.best_estimator_
clflasso = grid_search_lasso_p.best_estimator_
```

```python
# Creating the matrix of models 
Models_p = []
Models_p.append(('Linear Regression',clfLR))
Models_p.append(('Linear ridge',clfridge))
Models_p.append(('Linear Lasso',clflasso))
```

```python
n_iterations = 100
n_size = int(len(concrete_df2_norm_imp) * 0.5)

# run bootsrap
names = []
for names,models in Models_p:
    for i in range (n_iterations):
        X_trainp, X_testp, y_trainp, y_testp = train_test_split(X_poly,y,test_size = 0.3, random_state = None)
        #fit model 
        models.fit(X_trainp,y_trainp)
        #evaluate model
        score = models.score(X_testp,y_testp)
        statistic.append(score)
    #plot scores
    plt.hist(statistic)
    plt.show()
    #confidence intervals
    alpha = 0.95
    p = ((1-alpha)/2) * 100
    lower = max(0.0,np.percentile(statistic,p))
    p = (alpha + ((1-alpha)/2)) * 100
    upper = min(1.0,np.percentile(statistic,p))
    print('%.1f likelihood that the confidence interval of %s is  %.1f and %.1f which covers the true skill of the model' %(alpha*100,names,lower*100, upper*100))
    statistic = []
    print('*' *100)
```

- CONFIDENCE INTERVAL CALCULATION FOR OTHER REGRESSION MODELS

```python
# Initiating Models
clfLR = LinearRegression()
clfridge = grid_search_ridge.best_estimator_
clflasso = grid_search_lasso.best_estimator_
final_KNNR = gs_knnr.best_estimator_
final_DTR = grid_search_DTR.best_estimator_
final_RFR = gs_rfr.best_estimator_
final_SVR = gs_svr.best_estimator_
final_GBR = gs_gbr.best_estimator_
final_XGBR = gs_xgbr.best_estimator_
```

```python

n_iterations = 100
n_size = int(len(concrete_df2_norm_imp) * 0.5)

# run bootsrap
statistic = []
names = []
lower = []
upper = []
for names,models in Models:
    for i in range (n_iterations):
        #prepare train and test sets
        X_train, X_test, y_train, y_test = train_test_split(X_z1,y,test_size = 0.3, random_state = None)
        #fit model 
        models.fit(X_train,y_train)
        #evaluate model
        score = models.score(X_test,y_test)
        statistic.append(score)
    #plot scores
    plt.hist(statistic)
    plt.show()
    #confidence intervals
    alpha = 0.95
    p = ((1-alpha)/2) * 100
    lowrange = max(0.0,np.percentile(statistic,p))
    lower.append(lowrange)
    p = (alpha + ((1-alpha)/2)) * 100
    highrange = min(1.0,np.percentile(statistic,p))
    upper.append(highrange)
    msg = " The accuracy confidence interval of %s model lies between %.1f and %.1f with %.1f likelihood " %(names,lowrange*100, highrange*100, alpha*100)
    print(msg)
    statistic = []
    print('*' *100)
```
